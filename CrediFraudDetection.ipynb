# **Preprocessing**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
train_df['trans_date_trans_time'] = pd.to_datetime(train_df['trans_date_trans_time'])
```

```python
train_df
```

```python
test_df = pd.read_csv('/content/fraudTest.csv')
```

```python
test_df['trans_date_trans_time'] = pd.to_datetime(test_df['trans_date_trans_time'])
```

```python
# Extract time-based features
train_df['hour'] = train_df['trans_date_trans_time'].dt.hour
train_df['day_of_week'] = train_df['trans_date_trans_time'].dt.dayofweek


test_df['hour'] = test_df['trans_date_trans_time'].dt.hour
test_df['day_of_week'] = test_df['trans_date_trans_time'].dt.dayofweek
```

```python
!pip install haversine
```

```python
from haversine import haversine
```

```python
def calculate_distance(row):
    user_location = (row['lat'], row['long'])
    merch_location = (row['merch_lat'], row['merch_long'])
    return haversine(user_location, merch_location)

train_df['distance'] = train_df.apply(calculate_distance, axis=1)
```

```python
test_df['distance'] = test_df.apply(calculate_distance, axis=1)
```

```python
train_df['dob'] = pd.to_datetime(train_df['dob'])
train_df['age'] = (pd.to_datetime('today') - train_df['dob']).dt.days // 365
```

```python
test_df['dob'] = pd.to_datetime(test_df['dob'])
test_df['age'] = (pd.to_datetime('today') - test_df['dob']).dt.days // 365
```

```python
train_df.drop(['Unnamed: 0', 'trans_num', 'trans_date_trans_time', 'cc_num', 'first',
         'last', 'street', 'city', 'state', 'zip', 'dob'], axis=1, inplace=True)
```

```python
test_df.drop(['Unnamed: 0', 'trans_num', 'trans_date_trans_time', 'cc_num', 'first',
         'last', 'street', 'city', 'state', 'zip', 'dob'], axis=1, inplace=True)
```

```python
train_df
```

```python
train_df = pd.get_dummies(train_df, columns=['category'], drop_first=True)
```

```python
train_df
```

```python
test_df = pd.get_dummies(test_df, columns=['category'], drop_first=True)
```

```python
X_train = train_df.drop(columns=['is_fraud','merchant','job'], axis=1)
y_train = train_df['is_fraud']
```

```python
X_train
```

```python
y_train
```

```python
X_test = test_df.drop(columns=['is_fraud','merchant','job'], axis=1)
y_test = test_df['is_fraud']
```

# **Model**

```python
# Scale numerical features
scaler = StandardScaler()

# Identify numerical columns
numerical_columns = ['amt', 'lat', 'long', 'merch_lat', 'merch_long', 'city_pop', 'distance', 'hour', 'day_of_week', 'age']

# Fit the scaler
X_train[numerical_columns] = scaler.fit_transform(X_train[numerical_columns])
X_test[numerical_columns] = scaler.transform(X_test[numerical_columns])
```

```python
X_train.columns
```

```python
X_train = pd.get_dummies(X_train, columns=['gender'], drop_first=True)
X_test = pd.get_dummies(X_test, columns=['gender'], drop_first=True)
```

```python
# Initialize the model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight="balanced")

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)
```

```python
y_pred
```

```python
# Print evaluation metrics
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# **Reading file**

```python
import pandas as pd
train_df = pd.read_csv('/content/Genre Classification Dataset/train_data.txt',sep=':::',names = ['','Title','Genre','Description'])
train_df
```

```python
test_df = pd.read_csv('/content/Genre Classification Dataset/test_data_solution.txt',sep=':::',names = ['Title','Genre','Description'])
test_df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_df['Genre'] = le.fit_transform(train_df['Genre'])
train_df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
test_df['Genre'] = le.fit_transform(test_df['Genre'])
test_df
```

```python
le.classes_
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python

train_df['Description'] = train_df['Description'].apply(word_tokenize)
train_df
```

```python
test_df['Description'] = test_df['Description'].apply(word_tokenize)
test_df
```

```python
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
stemmer = SnowballStemmer("english")
stop_words = set(stopwords.words('english'))
def stem_text(text):

    # Remove stopwords and apply stemming
    return [stemmer.stem(word) for word in text if word.lower() not in stop_words]
   # Join the stemmed words back into a sentence

# Step 5: Apply stemming and stopword removal to the dataset
train_df['Description'] = train_df['Description'].apply(stem_text)
test_df['Description'] = test_df['Description'].apply(stem_text)
train_df
test_df
```

```python
train_df
```

```python
X_train = train_df['Description']
y_train = train_df['Genre']
X_test = test_df['Description']
y_test = test_df['Genre']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=5000,stop_words='english')
X_train_str = [' '.join(tokens) for tokens in X_train]
X_test_str = [' '.join(tokens) for tokens in X_test]
X_train = vectorizer.fit_transform(X_train_str)
X_test = vectorizer.transform(X_test_str)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(class_weight='balanced')
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
